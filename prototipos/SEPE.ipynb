```python
import pandas as pd
```

# Paro por municipio

```python
url_base = 'https://sede.sepe.gob.es/es/portaltrabaja/resources/sede/datos_abiertos/datos/Paro_por_municipios_{}_csv.csv'

anios = range(2006, 2019)
anio = 2018
```

```python
url = url_base.format(anio)

df = pd.read_csv(url, sep=';', encoding='latin1', header=1)

# Elimina columnas innecesarias
columnas = ['Código mes ', 'Código de CA', 'Codigo Provincia', 'Codigo Municipio']
df.drop(columnas, axis=1, inplace=True)

df.head()
```
