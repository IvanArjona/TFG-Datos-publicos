```python
import requests
import pandas as pd
```

```python
url = "http://servicios.ine.es/wstempus/js/es/DATOS_TABLA/2852?"

respuesta = requests.get(url)
datos = respuesta.json()
```

```python
df_rows = []

for row in datos:
    fechas = pd.to_datetime([data['Fecha'] for data in row['Data']], unit='ms', utc=True)
    fechas = fechas.tz_convert('Europe/Madrid').tz_localize(None)
    
    poblaciones = [x['Valor'] for x in row['Data']]
    campos = row["Nombre"].split(". ")
    
    ambito = campos[0]
    genero = campos[1]
    medida = '{} ({})'.format(campos[2], campos[3])
    
    data = {
        'Ambito': ambito,
        'Genero': genero,
        'Fecha': fechas,
        medida: poblaciones
    }
    
    df_row = pd.DataFrame(data)
    df_rows.append(df_row)

df = pd.concat(df_rows)

df.head()
```

```python
df = pd.pivot_table(df, index=['Ambito', 'Fecha'], columns=['Genero'])

df.head()
```

```python
df.reset_index(inplace=True)
df.columns = ['Ambito', 'fecha',' Hombres', 'Mujeres', 'Total']

df.head()
```
